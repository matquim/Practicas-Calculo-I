# Límites y continuidad

## Práctica 3: Continuidad

````{admonition} Concepto
:class: tip

Una función es continua en $x=a$ si:

a) $f(a)$ está definida

b) $\displaystyle{\lim_{x \to a} f(x)}$ existe

c) $\displaystyle{\lim_{x \to a} f(x) = f(a)}$

````


**Ejemplo**


`````python
def f(x):
    return x**2

print(f(2))
`````
