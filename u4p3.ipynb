## Práctica 3: Interpretación geométrica

**Concepto teórico**

La derivada representa la pendiente de la recta tangente a la curva en un punto.

**Ejemplo**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

x0 = 1
m = 2*x0  # derivada exacta

x = np.linspace(-1, 3, 100)
y = f(x)

recta = m*(x - x0) + f(x0)

plt.plot(x, y)
plt.plot(x, recta)
plt.scatter([x0], [f(x0)])
plt.show()
```
