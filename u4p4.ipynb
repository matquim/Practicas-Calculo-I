# Límites y continuidad

## Práctica 4: Visualización

**Objetivo**

Visualizar funciones para interpretar su comportamiento.

**Ejemplo**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 100)
y = (x**2 - 1)/(x - 1)

plt.plot(x, y)
plt.axvline(1)
plt.show()
```
