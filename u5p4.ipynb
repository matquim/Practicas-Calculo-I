La **Ley de Enfriamiento de Newton** describe cómo la temperatura $T(t)$ de un objeto cambia con el tiempo cuando está en contacto con un medio ambiente a temperatura constante $T_{\text{amb}}$. La ecuación diferencial que modela este fenómeno es:

\begin{equation}
\frac{dT}{dt} = -k (T - T_{\text{amb}})
\end{equation}

Donde:
- $T(t)$: Temperatura del objeto en el tiempo $t$.
- $T_{\text{amb}}$: Temperatura ambiente (constante).
- $k$: Constante de enfriamiento (depende del material y condiciones del medio).

**Solución analítica:**
\begin{equation}
T(t) = T_{\text{amb}} + (T_0 - T_{\text{amb}}) e^{-kt}
\end{equation}
Donde $ T_0 $ es la temperatura inicial del objeto en $t = 0$.

Instrucciones

1. Derivar la solución analítica de la Ley de Enfriamiento de Newton a partir de su ecuación diferencial.
2. Calcular la derivada $\frac{dT}{dt}$ y analizar su comportamiento en función del tiempo.
3. Determinar el tiempo $t$ en el que la temperatura del objeto es igual a la temperatura promedio entre $T_0$ y $T_{\text{amb}}$.
4. Graficar $T(t)$ y $\frac{dT}{dt}$ en función del tiempo.
5. Interpretar físicamente los resultados, incluyendo:
   - ¿Qué representa la constante $k$ en términos físicos?
   - ¿Cómo afecta el valor de $k$ a la velocidad de enfriamiento?
   - ¿Qué sugiere el signo de $\frac{dT}{dt}$ sobre la dirección del flujo de calor?

**Solución**

**Derivación de la Solución Analítica**
La ecuación diferencial de la Ley de Enfriamiento de Newton es:
\begin{equation}
\frac{dT}{dt} = -k (T - T_{\text{amb}})
\end{equation}

1. Separar variables:
  \begin{equation}
   \frac{dT}{T - T_{\text{amb}}} = -k \, dt
   \end{equation}

2. Integrar ambos lados:
  \begin{equation}
   \int \frac{1}{T - T_{\text{amb}}} \, dT = -k \int dt
   \end{equation}
  \begin{equation}
   \ln T - T_{\text{amb}}| = -kt + C
   \end{equation}

3. Exponenciar ambos lados:
  \begin{equation}
   T - T_{\text{amb}} = e^{-kt + C} = e^C e^{-kt}
  \end{equation}
   Usando la condición inicial $T(0) = T_0$, obtenemos $e^C = T_0 - T_{\text{amb}}$:
  \begin{equation}
   T(t) = T_{\text{amb}} + (T_0 - T_{\text{amb}}) e^{-kt}
  \end{equation}

**Resultado:**
\begin{equation}
\boxed{T(t) = T_{\text{amb}} + (T_0 - T_{\text{amb}}) e^{-kt}}
\end{equation}

**Derivada de $T(t)$ con Respecto al Tiempo**
Derivamos la solución analítica:
\begin{equation}
\frac{dT}{dt} = \frac{d}{dt} \left[ T_{\text{amb}} + (T_0 - T_{\text{amb}}) e^{-kt} \right] = -k (T_0 - T_{\text{amb}}) e^{-kt}
\end{equation}

**Resultado:**
\begin{equation}
\boxed{\frac{dT}{dt} = -k (T_0 - T_{\text{amb}}) e^{-kt}}
\end{equation}

**Interpretación:**
- La derivada $\frac{dT}{dt}$ es **negativa** si $T_0 > T_{\text{amb}}$ (el objeto se enfría).
- La magnitud de $\frac{dT}{dt}$ **disminuye exponencialmente** con el tiempo.

**Tiempo en el que $T(t)$ es Igual a la Temperatura Promedio**

La temperatura promedio entre $T_0$ y $T_{\text{amb}}$ es:
\begin{equation}
T_{\text{prom}} = \frac{T_0 + T_{\text{amb}}}{2}
\end{equation}

Igualamos $T(t) = T_{\text{prom}}$:
\begin{equation}
T_{\text{amb}} + (T_0 - T_{\text{amb}}) e^{-kt} = \frac{T_0 + T_{\text{amb}}}{2}
\end{equation}

**Solución:**
1. Restar $T_{\text{amb}}$ a ambos lados:
  \begin{equation}
   (T_0 - T_{\text{amb}}) e^{-kt} = \frac{T_0 - T_{\text{amb}}}{2}
   \end{equation}

2. Dividir entre $T_0 - T_{\text{amb}}$:
  \begin{equation}
   e^{-kt} = \frac{1}{2}
   \end{equation}

3. Aplicar logaritmo natural:
  \begin{equation}
   -kt = \ln \left( \frac{1}{2} \right) = -\ln 2 \implies t = \frac{\ln 2}{k}
   \end{equation}

**Resultado:**
\begin{equation}
\boxed{t = \frac{\ln 2}{k}}
\end{equation}

**Interpretación:**
- Este tiempo es **independiente de $T_0$ y $T_{\text{amb}}$** y solo depende de $k$.
- Es el tiempo en el que el objeto ha reducido su diferencia de temperatura con el ambiente a la mitad.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
T0 = 100.0       # Temperatura inicial del objeto (°C)
T_amb = 20.0     # Temperatura ambiente (°C)
k = 0.1          # Constante de enfriamiento (min^-1)

# Tiempo
t = np.linspace(0, 50, 500)  # Rango de tiempo (min)

# Temperatura T(t)
T = T_amb + (T0 - T_amb) * np.exp(-k * t)

# Derivada dT/dt
dT_dt = -k * (T0 - T_amb) * np.exp(-k * t)

# Temperatura promedio
T_prom = (T0 + T_amb) / 2

# Tiempo en el que T(t) = T_prom
t_prom = np.log(2) / k

# Gráfica
plt.figure(figsize=(12, 6))

# Gráfica de T(t)
plt.subplot(1, 2, 1)
plt.plot(t, T, label="$T(t)$", color="blue")
plt.axhline(T_amb, color="red", linestyle="--", label="$T_{\\text{amb}}$")
plt.axhline(T_prom, color="green", linestyle="--", label="$T_{\\text{prom}}$")
plt.axvline(t_prom, color="purple", linestyle=":", label="$t = \\frac{\\ln 2}{k}$")
plt.xlabel("Tiempo (min)")
plt.ylabel("Temperatura (°C)")
plt.title("Ley de Enfriamiento de Newton: $T(t)$")
plt.legend()
plt.grid(True)

# Gráfica de dT/dt
plt.subplot(1, 2, 2)
plt.plot(t, dT_dt, label="$\\frac{dT}{dt}$", color="orange")
plt.axhline(0, color="black", linestyle="--")
plt.xlabel("Tiempo (min)")
plt.ylabel("Velocidad de cambio de temperatura (°C/min)")
plt.title("Derivada de $T(t)$")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Decaimiento Radiactivo

## 📚 Contexto Teórico
El **decaimiento radiactivo** describe cómo la cantidad de un **isótopo radiactivo** disminuye con el tiempo debido a la emisión de partículas o radiación. La ecuación diferencial que modela este fenómeno es:

\begin{equation}
\frac{dN}{dt} = -\lambda N
\end{equation}

Donde:
- $N(t)$: Número de núcleos radiactivos en el tiempo $t$.
- $\lambda$: Constante de decaimiento (probabilidad de decaimiento por unidad de tiempo).

**Solución analítica:**
\begin{equation}
N(t) = N_0 e^{-\lambda t}
\end{equation}
Donde $N_0$ es el número inicial de núcleos radiactivos en $t = 0$.

**Tiempo de vida media ($t_{1/2}$):**
\begin{equation}
t_{1/2} = \frac{\ln 2}{\lambda}
\end{equation}

## 📝 Solución

---

### 1. Derivación de la Solución Analítica
La ecuación diferencial del decaimiento radiactivo es:
\begin{equation}
\frac{dN}{dt} = -\lambda N
\end{equation}

**Solución:**
1. Separar variables:
  \begin{equation}
   \frac{dN}{N} = -\lambda \, dt
   \end{equation}

2. Integrar ambos lados:
  \begin{equation}
   \int \frac{1}{N} \, dN = -\lambda \int dt
   \end{equation}
  \begin{equation}
   \ln|N| = -\lambda t + C
   \end{equation}

3. Exponenciar ambos lados:
  \begin{equation}
   N = e^{-\lambda t + C} = e^C e^{-\lambda t}
   \end{equation}
   Usando la condición inicial $N(0) = N_0$, obtenemos $e^C = N_0$:
  \begin{equation}
   N(t) = N_0 e^{-\lambda t}
   \end{equation}

**Resultado:**
\begin{equation}
\boxed{N(t) = N_0 e^{-\lambda t}}
\end{equation}

---

### 2. Derivada de $N(t)$ con Respecto al Tiempo
Derivamos la solución analítica:
\begin{equation}
\frac{dN}{dt} = \frac{d}{dt} \left[ N_0 e^{-\lambda t} \right] = -\lambda N_0 e^{-\lambda t}
\end{equation}

**Resultado:**
\begin{equation}
\boxed{\frac{dN}{dt} = -\lambda N_0 e^{-\lambda t}}
\end{equation}

**Interpretación:**
- La derivada $\frac{dN}{dt}$ es **negativa** para todo $t$, lo que indica que $N$ **siempre disminuye** con el tiempo.
- La **velocidad de decaimiento** $\left| \frac{dN}{dt} \right| = \lambda N_0 e^{-\lambda t}$ es **máxima en $t = 0$** y **disminuye exponencialmente** con el tiempo.

---

### 3. Tiempo en el que la Velocidad de Decaimiento es Máxima y Mínima
- **Máxima velocidad de decaimiento**:
 \begin{equation}
  \left| \frac{dN}{dt} \right|_{t=0} = \lambda N_0
  \end{equation}

- **Mínima velocidad de decaimiento**:
 \begin{equation}
  \lim_{t \to \infty} \left| \frac{dN}{dt} \right| = 0
  \end{equation}

**Interpretación:**
- Al inicio ($t = 0$), la velocidad de decaimiento es **máxima** porque hay más núcleos radiactivos disponibles para decaer.
- Con el tiempo, la velocidad de decaimiento **disminuye** porque hay menos núcleos radiactivos.

---

### 4. Actividad Radiactiva y su Derivada
La **actividad radiactiva** $A(t)$ se define como el número de decaimientos por unidad de tiempo:
\begin{equation}
A(t) = \lambda N(t) = \lambda N_0 e^{-\lambda t}
\end{equation}

**Derivada de $A(t)$**:
\begin{equation}
\frac{dA}{dt} = \frac{d}{dt} \left[ \lambda N_0 e^{-\lambda t} \right] = -\lambda^2 N_0 e^{-\lambda t}
\end{equation}

**Resultado:**
\begin{equation}
\boxed{A(t) = \lambda N_0 e^{-\lambda t}}
\end{equation}
\begin{equation}
\boxed{\frac{dA}{dt} = -\lambda^2 N_0 e^{-\lambda t}}
\end{equation}

**Interpretación:**
- La actividad radiactiva $A(t)$ **disminuye exponencialmente** con el tiempo.
- La derivada $\frac{dA}{dt}$ es **negativa**, lo que indica que la actividad **disminuye** con el tiempo.
