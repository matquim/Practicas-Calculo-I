````{admonition} Para Saber Más: Convertir un número real en un número flotante
:class: note, dropdown

Probar la función `bin` con un número de coma flotante (ej. `bin(5.4)`) conducirá directamente a un error.

Convertir un número real a uno flotante (según el estándar IEEE 754) para su almacenamiento en la computadora consiste en determinar los siguientes elementos:
1. El **signo**, llamado $s$. Indica si es un número positivo o negativo, éste se almacena en *1 bit*.
2. Un **exponente** , llamado $e$. Es una  potencia de 2, el cual se almacena en *11 bits*.
3. La **mantisa** $f$. Es el coeficiente del exponente y se almacena en *52 bits*.

Estos se usan para representar el número real en la forma:

\begin{equation*}
n_f = (-1)^s 2^{e-1023} \left( 1+f\right)
\end{equation*}

Por ejemplo, para el $5.75$ se haría lo siguiente:

1. Determinar el signo
- El número $5.75$ es positivo por lo que $s=0$.

2. Convertir la parte entera a binario
- El número $5.75$ tiene como parte entera $5$, que en binario es 101

3. Convertir la parte decimal a binario multiplica por 2 repetidamente y extrayendo la parte entera resultante:
- $0.75 \times 2 = 1.5$
- $0.5  \times 2 = 1.0$
    - Por lo que el número $0.75$ en binario es $0.11$.

**Conclusión:** El número $5.75$ en binario es $101.11$

4. Normalizar el número escribiéndolo como $1.(\text{mantisa}) \times 2^{\text{exponente}}$.
- El número $101.11$ se convierte en $1.0111\times 2^{2}$
- Como el punto decimal se movió dos lugares, el exponente es $2$
- La mantisa es $0111$

5. Calcular el exponente con sesgo, mediante la fórmula $e = \text{exponente} + 1023$
- Convertir el exponente ajustado al sego en representación binaria de 11 bits.
- Si la representación binaria no ocupa 11 bits, se rellena con ceros.
    - El exponente de $1.0111 \times 2^2$ es dos, por lo que el exponente segado está dado por: $e= 2+1023=1025$
    - El número $1025$ en binario es $10000000001$. Como ocupa 11 bits no es necesario completar con ceros.*

6. Tomar la parte fraccionaria de la representaciónn binaria normalizada, es decir lo que queda después del punto decimal y ajustar la mantisa para completar 52 bits. Si son menos, rellenar con ceros.

- La parte fraccionaria del número $5.75$ es $0111$ por lo que la mantisa con 52 bits es: $0111000000000000000000000000000000000000000000000000$

7. La representación del número 5.75 consiste en 64 bits distribuídos de la siguiente forma:

$$
\begin{aligned}
0 &  & 10000000001 & & 0111000000000\\
  &  &             & & 0000000000000\\
  &  &             & & 0000000000000\\
  &  &             & & 0000000000000
\end{aligned}
$$

Se proporciona el siguiente código para ver la representación binaria de un número de coma flotante:

~~~python
import struct

x = 5.75

packed_x = struct.pack('>d', x)
int_rep = int.from_bytes(packed_x)
binario_flotante = bin(int_rep)

print(f"El número {x} en binario (IEEE 754, 64 bits) es:\n{binario_flotante}")
~~~
````

El hecho de que se use un número determinado de bits para representar un número con decimales implica que la precisión es finita. Es decir, no todo número puede ser representado con los 64 bits y solo podemos usar una cantidad limitada de decimales. Esto quiere decir que se designan un total de 64 bits para almacenar cada uno de ellos. 

Los números de punto flotante son una aproximación computacional a los números reales. Cualquier número que requiera más decimales o cuya representación requiera más bits de los disponibles, será truncado, introduciendo un error por precisión numérica.

Por ejemplo, la representación exacta del número  $\sqrt{2}$ requiere que almacenemos una cantidad infinita de números decimales, lo mismo que los números $\frac{1}{3}= 0.\bar{3}$, $\frac{1}{7}= 0.\overline{142857}$, etc.

En Python, la información de un número flotante se obtiene usando el paquete `sys`.

`````python
import sys
sys.float_info
`````

Como puede ver, Python es capaz de almacenar números muy grandes y con una gran cantidad de decimales. Si bien el límite no es infinito, es lo suficiente amplio para que lo usemos y no nos volvamos a preocupar por esto.

Sin embargo, es importante que conozca que todo tiene un límite y que esto puede volverse importante en situaciones más críticas más avanzadas, desde la trayectoria de un satélite hasta las ecuaciones diferenciales que modelan el comportamiento de un reactor, la solución de la química cuántica y la acumulación de errores en simulaciones computacionales de dinámica molecular.

4. El siguiente código compara dos números flotantes, modifícalo y ejecútalo para comparar el resultado aritmético de las operaciones realizadas en el ejercicio 3 con el resultado numérico de dichas operaciones con Python.

`````python
# Actividad 2
import struct

def calculo_representacion_flotante(x):
    # Empaquetar el número como flotante de 64 bits
    packed_x = struct.pack('>d', x)
    # Convertir a un entero de 64 bits para ver los bits
    int_rep = int.from_bytes(packed_x, byteorder='big')
    # Convertir a binario
    binario_flotante = bin(int_rep)[2:].zfill(64)
    return binario_flotante

numero_real = 0.1
binario_flotante = calculo_representacion_flotante(numero_real)
print(f"El número {numero_real} en binario (IEEE 754, 64 bits) es:\n{binario_flotante}")
`````
