# Trabajo en una Expansión Isotérmica

````{admonition} Concepto
:class: tip


En termodinámica, el **trabajo $ W $** realizado por un gas durante una expansión o compresión se calcula como:

\begin{equation}
W = \int_{V_i}^{V_f} P \, dV
\end{equation}

si la temperatura se mantiene constante, esto se conoce como el `trabajo en una expansión isotérmica`.
````

Para un **gas ideal**, la ecuación de estado es:

$$
P = \frac{n R T}{V}
$$

Donde:
- $ P $: Presión del gas.
- $ V $: Volumen del gas.
- $ n $: Número de moles.
- $ R $: Constante de los gases ($8.314 \, \text{J/mol·K}$).
- $ T $: Temperatura (constante en un proceso isotérmico).

Considere un gas ideal a temperatura $T = 300 \, \text{K}$, que se expande desde un volumen $ V_i = 1 \, \text{L} $ hasta $ V_f = 5 \, \text{L} $.

Realice la gráfica $P$ vs #V$ para $n = 2 moles$ y sombree el área bajo la curva (que representa el trabajo durante la expansión).

`````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

n = 2.0       # Moles
R = 8.314     # J/mol·K
T = 300.0     # K
Vi = 1.0      # L (volumen inicial)
Vf = 5.0      # L (volumen final)

V = np.linspace(Vi, Vf, 500)
P = n * R * T / V

plt.figure(figsize=(10, 6))

plt.plot(V, P, label="$P(V) = \\frac{nRT}{V}$", color="blue")
plt.fill_between(V, P, alpha=0.3, color="blue", label="Área = Trabajo $W$")

plt.xlabel("Volumen $V$ (L)")
plt.ylabel("Presión $P$ (Pa)")
plt.title("Trabajo en una Expansión Isotérmica de un Gas Ideal")

plt.legend()
plt.grid(True)
`````

Utilice la función `scipy.quad` para integrar numéricamente el área bajo la  curva

`````python
W, _ = quad(lambda V: n * R * T / V, Vi, Vf)
print(f"Trabajo W = {W:.2f} J")
`````

Para el gas ideal, la expresión analítica para el trabajo está dada por:

\begin{equation}
W = n R T \ln \left( \frac{V_f}{V_i} \right)
\end{equation}

Calcule el trabajo usando esta expresión, y compárelo con el resultado obtenido mediante integración numérica.

`````python
W = n * R* T * np.log(Vf / Vi)
print(f"Trabajo W = {W:.2f} J")
`````
