## Decaimiento Radiactivo

````{admonition} Concepto
:class: tip

Los isótopos son átomos de un mismo elemento que sólo difieren en la cantidad de neutrones en el núcleo. Muchos de estos casos son radioactivos, es decir, que con el tiempo emiten partículas del núcleo para convertirse en átomos más estables. Este fenómeno se conoce como decaimiento radioactivo, y se modela mediante la ecuación:

\begin{equation}
\frac{dN}{dt} = -\lambda N
\end{equation}

Donde:
- $N(t)$: Número de núcleos radiactivos en el tiempo $t$.
- $\lambda$: Constante de decaimiento (probabilidad de decaimiento por unidad de tiempo).

````

En la práctica, un instrumento mide la masa de los núcleos radioactivos.

Los colegas del laboratorio tomaron un contador Geiger y midieron la masa de Iodo-131 radioactivo a varios tiempos, obteniendo los siguientes datos:

t = [0.0, 1.0, 2.0, 3.5, 5.0, 6.5, 8.0, 10.0, 12.0, 14.0, 16.0, 19.0, 22.0, 25.0, 28.0, 32.0]
~~~

Masa radioactiva (mg)
~~~
N = [100.0, 91.8, 84.2, 74.0, 65.0, 57.1, 50.1, 42.1, 35.4, 29.8, 25.0, 19.2, 14.8, 11.4, 8.8, 6.2]
~~~

Grafique el número de partículas en función del tiempo

`````python
import matplotlib.pyplot as plt
import numpy as np

t = np.array([0.0, 1.0, 2.0, 3.5, 5.0, 6.5, 8.0, 10.0, 12.0, 14.0, 16.0, 19.0, 22.0, 25.0, 28.0, 32.0])
N = np.array([100.0, 91.8, 84.2, 74.0, 65.0, 57.1, 50.1, 42.1, 35.4, 29.8, 25.0, 19.2, 14.8, 11.4, 8.8, 6.2])

plt.plot(t, N, "o-")
`````

Evalúe numéricamente la velocidad de decaimiento:

\begin{equation}
\frac{dN}{dt} \approx \frac{N_{i+1} - N_{i}}{t_{i+1} - t_{i}}
\end{equation}

grafíque la derivada en función del tiempo.

`````python
dN_dt = (N[1:] - N[0:-1]) / (t[1:] - t[0:-1])
tm = (t[0:-1] + t[1:]) / 2

plt.figure(figsize=(8, 5))

plt.plot(tm, dN_dt, marker='o', color='teal', label='Primera derivada')
    
plt.xlabel('Tiempo (min)')
plt.ylabel('dN/dt (C/min)')

plt.grid(True)

plt.legend()
plt.show()
`````

Este fenómeno se modela mediante
\begin{equation}
\frac{dN}{dt} = -\lambda N
\end{equation}

Realice la gráfica correspondiente y obtenga la constante de decaimiento mediante una regresión lineal.

`````python
Nm = (N[0:-1] + N[1:]) / 2

plt.figure(figsize=(8, 5))

plt.plot(Nm, dN_dt, marker='o', color='teal', label='Primera derivada')
    
plt.xlabel('Temperatura (C)')
plt.ylabel('dN/dt (C/min)')

plt.grid(True)

plt.legend()
plt.show()
`````

`````python
m, b = np.polyfit(Nm, dN_dt, 1)

l = -m

print(f"La constante de enfriamiento aproximada l es: {l:.4f} nucleos/min^-1")
`````
