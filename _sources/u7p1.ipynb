En termodinámica, el **trabajo $ W $** realizado por un gas durante una expansión o compresión se calcula como:

$$
W = \int_{V_1}^{V_2} P \, dV
$$

Para un **gas ideal**, la ecuación de estado es:

$$
P = \frac{n R T}{V}
$$

Donde:
- $ P $: Presión del gas.
- $ V $: Volumen del gas.
- $ n $: Número de moles.
- $ R $: Constante de los gases ($ 8.314 \, \text{J/mol·K} $).
- $ T $: Temperatura (constante en un proceso isotérmico).

**Trabajo en una expansión isotérmica**:

$$
W = n R T \ln \left( \frac{V_2}{V_1} \right)
$$

**Instrucciones**

1. Un gas ideal a temperatura $ T = 300 \, \text{K} $ se expande isotérmicamente desde un volumen $ V_1 = 1 \, \text{L} $ hasta $ V_2 = 5 \, \text{L} $. Calcular el **trabajo $ W $** realizado por el gas si $ n = 2 \, \text{moles} $.
2. **Graficar** la presión $ P $ vs. volumen $ V $ durante la expansión y el área bajo la curva (que representa el trabajo).
3. **Interpretar físicamente**:
   - ¿Por qué el trabajo es **positivo** en una expansión?
   - ¿Cómo cambiaría el trabajo si la expansión fuera **adiabática**?



Solución

**1. Cálculo del Trabajo $ W $**
Usamos la integral para el trabajo en una expansión isotérmica:
$$
W = n R T \ln \left( \frac{V_2}{V_1} \right)
$$
Sustituyendo los valores:
$$
W = 2 \, \text{mol} \times 8.314 \, \text{J/mol·K} \times 300 \, \text{K} \times \ln \left( \frac{5}{1} \right) \approx 8037.6 \, \text{J}
$$

**Resultado:**
$$
W \approx 8037.6 \, \text{J}
$$

`````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Parámetros
n = 2.0       # Moles
R = 8.314     # J/mol·K
T = 300.0     # K
V1 = 1.0      # L (volumen inicial)
V2 = 5.0      # L (volumen final)

# Volúmenes para la gráfica
V = np.linspace(V1, V2, 500)

# Presión P(V) = nRT / V
P = n * R * T / V

# Trabajo W (área bajo la curva)
W, _ = quad(lambda V: n * R * T / V, V1, V2)
print(f"Trabajo W = {W:.2f} J")

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(V, P, label="$P(V) = \\frac{nRT}{V}$", color="blue")
plt.fill_between(V, P, alpha=0.3, color="blue", label="Área = Trabajo $W$")
plt.xlabel("Volumen $V$ (L)")
plt.ylabel("Presión $P$ (Pa)")
plt.title("Trabajo en una Expansión Isotérmica de un Gas Ideal")
plt.legend()
plt.grid(True)
plt.show()
`````
