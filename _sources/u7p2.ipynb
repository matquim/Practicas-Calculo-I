# Masa Total a partir de una Distribución de Densidad

````{admonition} Concepto
:class: tip

La densidad de masa lineal modela como varía la masa de un cuerpo por unidad de longitud. Si tenemos una función para la densidad de masa, entonces podemos obtener la masa del cuerpo en el intervalo de longotid $[a, b]$ mediante la integral:

\begin{equation}
M = \int_{a}^{b} \lambda(x) dx
\end{equation}
````

Considere una barra de 20 cm de un material cuya densidad de masa varía con la posición como: 
\begin{equation}
   \lambda(x) = \lambda_0 e^{-k x}
\end{equation}

donde $ \lambda_0 = 0.5 \, \text{g/cm} $ y $ k = 0.1 \, \text{cm}^{-1} $.

Realice la gráfica de $x$ vs $\lamda(x)$ y sombree el área bajo la curva (que representa la masa del material).

`````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

lambda_0 = 0.5  # g/cm
k = 0.1         # cm^-1
L = 20.0        # cm

x = np.linspace(0, L, 500)

lambda_x = lambda_0 * np.exp(-k * x)

plt.figure(figsize=(10, 6))

plt.plot(x, lambda_x, label="$\\lambda(x) = \\lambda_0 e^{-k x}$", color="blue")
plt.fill_between(x, lambda_x, alpha=0.3, color="blue", label="Área = Masa total $M$")

plt.xlabel("Posición $x$ (cm)")
plt.ylabel("Densidad lineal de masa $\\lambda(x)$ (g/cm)")
plt.title("Distribución de Densidad Lineal de Masa")

plt.legend()
plt.grid(True)
`````

Utilice `scipy.quad` para calcular el área bajo la curva, la cual representa la masa total de la barra.

`````python
M, _ = quad(lambda x: lambda_0 * np.exp(-k * x), 0, L)
print(f"Masa total M = {M:.4f} g")
`````

Podemos realizar la integral analítica para obtener:

\begin{equation}
M = \int_{0}^{L} \lambda_0 e^{-k x} \, dx = \frac{\lambda_0}{k} \left( 1 - e^{-k L} \right)
\end{equation}

Sustituya los valores y obtenga el valor de la masa de la barra:

`````python
M = (0.5/0.1) * ( 1 - np.exp(-0.1 * 20) )
print(f"Masa total M = {M:.4f} g")
`````
