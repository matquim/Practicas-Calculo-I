# Sumas de Riemann

Las sumas de Riemann permiten aproximar el área bajo una curva usando rectángulos.
Supongamos que tenemos una función $f(x)$  y queremos calcular el área bajo la curva entre $x=a$ y $x=b$, tal como se muestra en la figura.

integracion.png

La idea será:

1. Dividir el intervalo $[a,b]$ en subintervalos del mismo tamaño (partición equidistante).
2. Construir una serie de rectángulos en cada subintervalo
3. Sumar las áreas de todos los rectángulos.


**Fórmula general:**

La suma de Riemann se calcula como:

\begin{align}
\displaystyle I^*=\sum_{i=1}^{n}{f(x_{i})\Delta x}
\end{align}

donde:
* $n ≡$ número de rectángulos
* $\Delta x ≡$ ancho de cada rectángulo (tamaño de la partición):

\begin{align}
\displaystyle \Delta x = \frac{b-a}{n}
\end{align}


* $f(x_i) = $ altura del rectángulo

barras.png

## El valor de $f(x_i)$

**Suma izquierda**

Su usa como altura de cada rectángulo el valor de la función en el extremo izquierdo de cada subintervalo.

**Suma derechas**

Su usa como altura de cada rectángulo el valor de la función en el extremo derecho de cada subintervalo.

**Suma punto medio**

Su usa como altura de cada rectángulo el valor de la función en el punto medio de cada subintervalo.

## Integral definida

Las sumas de Riemann son una base de la integral definida.

Cuando el número de rectángulos aumenta, la aproximáción de la suma de los rectángulos, se vuelve exactamente el área bajo la curva, es decir,

\begin{align}
\displaystyle \lim_{n\to \infty}{\sum_{i=i}^{n}{f(x_i)\Delta x}}=\int_{a}^{b}{f(x)\,dx}
\end{align}

## Metodología
Tomamos una partición de puntos de la forma:
\begin{align}
x_i=a+i\Delta x
\end{align}

en donde    

\begin{align}
\displaystyle \Delta x=\frac{b-a}{n}
\end{align}

La aproximación $I^*$  para la integral (suma de Riemann) está dada por:


\begin{align}
\displaystyle I^*=\sum_{i=0}^{n}{f(x_{i})h}
\end{align}

donde $x_i$ es el extremo izquierdo, derecho o el punto medio.

```python
# Importemos la biblioteca de Numpy
import numpy as np
```

```python
#Definamos la función
def f(x):
    return x**2
```

```python
#Definamos los extremos del intervalo a calcular la Suma de Riemann

a = 0
b = 2

#Definamos el número de subinetrvalos para establecer los rectángulos

n = 4

#Definamos el tamaño de cada subintervalo (tamaño de la partición)

dx = (b-a)/n

#Definamos todos los puntos de la partición

x = np.linspace(a,b,n+1)
print(x)
```

```python
#Definamos los puntos izquierdos
x_izq = x[0:n]
print(x_izq)

#Definamos los puntos derechos
x_der = x[1:n+1]
print(x_der)

#Definamos los puntos medios
x_med = (x_der + x_izq)/2
print(x_med)
```

```python
#Evaluemos la función en cada punto del intervalo
f(x_der)
```

```python
#Calculemos el área de cada rectángulo
A_izq = f(x_izq) * dx
print(A_izq)

A_der = f(x_der) * dx
print(A_der)

A_med = f(x_med) * dx
print(A_med)
```

```python
#Calculemos las sumas de Riemann:izquierdas, derechas y medias
Suma_izq = sum(A_izq)
Suma_der = sum(A_der)
Suma_med = sum(A_med)

print(f"Las suma de Riemann izquierda: I_izq = {Suma_izq}")
print(f"Las suma de Riemann derecha:   I_der = {Suma_der}")
print(f"Las suma de Riemann media:     I_med = {Suma_med}")
```

```python
# Ahora aumentemos en número de rectángulos
import numpy as np

def f(x):
    return x**2

#Extremos del intervalo
a = 0
b = 2

#Número de subintervalos
n = 1000

#Tamaño de la partición
dx = (b-a)/n

#Puntos de la partición
x = np.linspace(a,b,n+1)
#print(x)
x_izq = x[0:n]             #izquierdos
x_der = x[1:n+1]           #derechos
x_med = (x_der + x_izq)/2 #medios

#Areas
A_izq = f(x_izq) * dx #izquierdas
A_der = f(x_der) * dx #derechas
A_med = f(x_med) * dx #medias

#Sumas de Riemann
Suma_izq = sum(A_izq)
Suma_der = sum(A_der)
Suma_med = sum(A_med)

print(f"Las suma de Riemann izquierda: I_izq = {Suma_izq}")
print(f"Las suma de Riemann derecha:   I_der = {Suma_der}")
print(f"Las suma de Riemann media:     I_med = {Suma_med}")
```

## Integral indefinida

Se define la función $\displaystyle G(x) = \int_a^x{f(t)\,dt}$. Notamos que $G$ es una función en donde al variar el valor del límite superior de la integral, cambia el valor de la integral. Calculemos dicha función para varios ejemplos:



## $\displaystyle f(t) = t$

$\displaystyle G(x) = \int_0^x{t\,dt}$, con $x \in [0, 5]$

¿Qué función es?

```python
import numpy as np

def f(t):
    return t
x = np.linspace(0, 5, 50)
Gx = []
#integración de f(t)
for xi in x:
    #Extremos del intervalo
    a = 0
    b = xi
    #Número de subintervalos
    n = 100
    #Tamaño de la partición
    dt = (b-a)/n
    #Puntos de la partición
    t = np.linspace(a,b,n+1)
    t_med = (t[0:n] + t[1:n+1])/2 #t medios
    A_med = f(t_med) * dt #medias
    #Suma de Riemann
    Suma = sum(A_med)
    Gx.append(Suma)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,Gx)
plt.scatter(x,f(x))
```
