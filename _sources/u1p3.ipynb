## Orden

````{admonition} Concepto
:class: tip

Para cuales quiera dos números reales, $a$ y $b$ sólo se cumple cualquiera de las tres afirmaciones siguientes (solo una a la vez):

- $a<b$
- $a>b$
- $a=b$

En Python, esto se realiza de manera muy parecida, resultando un valor de verdad (True/False) para cada caso:
~~~
a < b
a > b
a == b
~~~
La única diferencia con la notación original es en la tercera expresión. La diferencia se debe a que en programación, `x=5` asigna el valor de 5 a la variable x, y `x==5` compara el valor de la variable `x` con el número 5 y revisa si son iguales.
````

Python puede determinar el orden de dos números reales, es decir, determinar el valor de verdad de afirmaciones que involucren propiedades de orden. En la siguiente celda de código, defina dos variables con dos números y evalúe el valor de las 3 afirmaciones:

`````python
a = 1.5
b = 2.5

print("a < b:", a < b)
print("a > b:", a > b)
print("a == b:", a == b)
`````

Dado que la comparación entre dos números regresa un valor de verdad, este puede usarse con la estructura `if`. Esta ejecuta selectivamente cierto fragmento de código sólo si (`if`) se cumple una condición, de lo contrario (`else`) se ejecuta otro fragmento de código. Pueden revisarse varios casos usando `elif`.

A continuación, se presenta un ejemplo para comparar dos números reales usando la estructura de `if`:

`````python
a = 1.5
b = 2.5

if a < b:
    print(f"El número {a} es menor que {b}")
elif a > b:
    print(f"El número {a} es mayor que {b}")
else:
    print(f"El número {a} es igual que {b}")
`````

## Valor absoluto
El valor absoluto de un número real es siempre un número positivo que geométricamente nos da la distancia de dicho número real al origen, se define como:

$
|x| = \left\{
  \begin{array}{lcc}
  -x & si & x < 0 \\
  x & si & x\geq 0 \\
      \end{array} \right.
$

Para calcular el valor absoluto de un un número real, ocupamos la función `abs()`. El siguiente código evalúa el valor absoluto de un núemero real.

`````python
numero_real = float(input("Ingresa un número real: "))

valor_absoluto = abs(numero_real)

print(f"El valor absoluto del número real {numero_real} es: {valor_absoluto}")
`````

## Valor absoluto de la diferencia de dos números reales.

````{admonition} Concepto
:class: tip
Dados dos números reales, $a$ y $b$, ocurre que

$
|a-b|=|b-a|
$

y además, $|a-b|$ resulta ser igual a la distancia en la recta real que hay entre los números $a$ y $b$.
````

 El siguiente código permite graficar dos números en la recta real y calcular el valor absoluto de la diferencia entre éstos. Observa que el valor absoluto de la resta es precisamente la distancia entre ambos números.

`````python
import matplotlib.pyplot as plt

# Ingresar los números directamente en lugar de utilizar input()
a = -3.0
b = -6.0

# Crear una figura y un eje
fig, ax = plt.subplots()

# Graficar los números en la recta real
ax.plot([a], [0], "r-o", label=f'a = {a}')
ax.plot([b], [0], 'b-o', label=f'b = {b}')

# Configuración de la recta real
ax.axhline(0, color='black', lw=1)
ax.set_ylim([-1, 1])
ax.set_xlabel('Recta Real')

# Etiquetas para los números
ax.text(a, 0.1, f'{a}', fontsize=12, ha='center')
ax.text(b, 0.1, f'{b}', fontsize=12, ha='center')

# Distancia: |a-b|
ax.text((a+b)/2, 0.3, f'|{a}-({b})|= {abs(a-b)}', fontsize=12, ha='center')

# Rango de la gráfica en el eje x
plt.xlim(min(a, b) - 1, max(a, b) + 1)

# Título y leyenda
plt.title('Representación de Números en la Recta Real')
plt.legend()

# Mostrar la gráfica
plt.show()
`````
