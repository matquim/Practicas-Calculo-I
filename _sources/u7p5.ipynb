**Contexto Teórico**
El **centro de masa** $ x_{\text{cm}} $ de un sistema de partículas se calcula como:
$$
x_{\text{cm}} = \frac{\int x \, \lambda(x) \, dx}{\int \lambda(x) \, dx}
$$
Donde $ \lambda(x) $ es la densidad lineal de masa.

**Enunciado**
1. Modelar la densidad lineal de masa $ \lambda(x) $ de una molécula de $ \text{CO} $ como dos funciones gaussianas centradas en $ x = -d/2 $ (carbono) y $ x = d/2 $ (oxígeno), donde $ d = 1.13 \, \text{Å} $:
   $$
   \lambda(x) = A \left( m_C e^{-\frac{(x + d/2)^2}{2 \sigma^2}} + m_O e^{-\frac{(x - d/2)^2}{2 \sigma^2}} \right)
   $$
   Donde $ A = 1 \, \text{u/Å} $, $ \sigma = 0.1 \, \text{Å} $, $ m_C = 12 \, \text{u} $, y $ m_O = 16 \, \text{u} $.
2. Calcular el **centro de masa $ x_{\text{cm}} $** de la molécula.
3. **Graficar** $ \lambda(x) $ y marcar la posición del centro de masa.
4. **Interpretar físicamente**:
   - ¿Por qué el centro de masa no está exactamente en $ x = 0 $?
   - ¿Cómo afecta la diferencia de masas atómicas al centro de masa?

**Solución**

**1. Cálculo del Centro de Masa $ x_{\text{cm}} $**
$$
x_{\text{cm}} = \frac{(m_O - m_C) d}{2 (m_C + m_O)} = \frac{(16 - 12) \times 1.13}{2 (12 + 16)} \approx 0.0807 \, \text{Å}
$$

**Resultado:**
$$
\boxed{x_{\text{cm}} \approx 0.0807 \, \text{Å}}
$$

`````python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
A = 1.0          # Amplitud (u/Å)
d = 1.13         # Distancia de enlace C-O (Å)
sigma = 0.1      # Ancho de los picos (Å)
m_C = 12.0       # Masa atómica del carbono (u)
m_O = 16.0       # Masa atómica del oxígeno (u)

# Posiciones x
x = np.linspace(-2, 2, 500)

# Densidad lineal de masa lambda(x)
lambda_x = A * (m_C * np.exp(-(x + d/2)**2 / (2 * sigma**2)) + m_O * np.exp(-(x - d/2)**2 / (2 * sigma**2)))

# Centro de masa x_cm
x_cm = (m_O - m_C) * d / (2 * (m_C + m_O))
print(f"Centro de masa x_cm = {x_cm:.4f} Å")

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, lambda_x, label="$\\lambda(x)$", color="blue")
plt.axvline(x_cm, color="red", linestyle="--", label=f"Centro de masa $x_{{cm}} = {x_cm:.4f}$ Å")
plt.axvline(-d/2, color="green", linestyle=":", label="Carbono ($x = -d/2$)")
plt.axvline(d/2, color="purple", linestyle=":", label="Oxígeno ($x = d/2$)")
plt.xlabel("Posición $x$ (Å)")
plt.ylabel("Densidad lineal de masa $\\lambda(x)$ (u/Å)")
plt.title("Distribución de Masa en la Molécula de CO")
plt.legend()
plt.grid(True)
plt.show()
`````
