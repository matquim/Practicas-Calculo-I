# Primera Ley de Fick

````{admonition} Concepto
:class: tip

La **primera ley de Fick** describe el flujo de difusión $ J $ de una sustancia en un medio:

\begin{equation}
J = -D \frac{dc}{dx}
\end{equation}

En un proceso de difusión en estado estacionario, la concentración $ c(x) $ de una sustancia puede modelarse como una distribución normal (gaussiana):

\begin{equation}
c(x) = \frac{M}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}}
\end{equation}

La probabilidad $ P $ de encontrar una partícula en el intervalo $[a, b]$ es proporcional a la integral de $ c(x) $ en ese intervalo:

\begin{equation}
P \propto \int_{a}^{b} c(x) \, dx
\end{equation}
````

Considere un soluto en una solución sigue una distribución normal con $ \mu = 0 \, \text{cm} $, $ \sigma = 1 \, \text{cm} $, y $ M = 1 \, \text{g} $.

Grafique $c(x)$ y el área bajo la curva en el intervalo $[-1, 1]$.

`````python
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt

M = 1.0       # Masa total (g)
sigma = 1.0   # Desviación estándar (cm)
mu = 0.0      # Posición media (cm)

def c(x):
    return (M / (sigma * np.sqrt(2 * np.pi))) * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Gráfica
x = np.linspace(-4, 4, 500)
c_x = c(x)

plt.figure(figsize=(10, 6))
plt.plot(x, c_x, label="$c(x)$", color="blue")
plt.fill_between(x, c_x, where=((x >= -1) & (x <= 1)), alpha=0.3, color="blue", label="Área = Probabilidad en $[-1, 1]$")
plt.xlabel("Posición $x$ (cm)")
plt.ylabel("Concentración $c(x)$ (g/cm)")
plt.title("Distribución de Concentración (Primera Ley de Fick)")
plt.legend()
plt.grid(True)
plt.show()
`````

Utilice `scipy.quad` para calcular el área bajo la curva, la cual representa la probabilidad relativa de encontrar el soluto en el intervalo $[-1, 1] \, \text{cm}$.

`````python
P, _ = quad(c, -1, 1)
print(f"Probabilidad relativa en [-1, 1] cm: {P:.4f}")
`````

Esto corresponde a realizar la integral:
$$
P \propto \int_{-1}^{1} \frac{M}{\sigma \sqrt{2 \pi}} e^{-\frac{x^2}{2 \sigma^2}} \, dx
$$

que, al sustituir los valores ($ M = 1 $, $ \sigma = 1 $) resulta en:

$$
P \propto \frac{1}{\sqrt{2 \pi}} \int_{-1}^{1} e^{-\frac{x^2}{2}} \, dx \approx 0.6827
$$

Es decir, la probabilidad es:

$$
P \approx 0.6827 \text{ (68.27\%)}
$$
