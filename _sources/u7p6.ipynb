# Trabajo en una Expansión Adiabática**

**Contexto Teórico**

En un **proceso adiabático** (sin transferencia de calor), la relación entre presión \( P \) y volumen \( V \) para un gas ideal está dada por:
\[
P V^\gamma = \text{constante}
\]
Donde \( \gamma = \frac{C_p}{C_v} \) es el **coeficiente adiabático**.

El **trabajo \( W \)** realizado por el gas durante una expansión adiabática es:
\[
W = \frac{P_1 V_1 - P_2 V_2}{\gamma - 1}
\]

**Enunciado**
1. Un gas ideal diatómico (\( \gamma = \frac{7}{5} \)) se expande adiabáticamente desde \( V_1 = 1 \, \text{L} \) y \( P_1 = 10 \, \text{atm} \) hasta \( V_2 = 4 \, \text{L} \). Calcular el **trabajo \( W \)** realizado por el gas.
2. **Graficar** \( P \) vs. \( V \) durante la expansión adiabática y el área bajo la curva (que representa el trabajo).
3. **Comparar** el trabajo adiabático con el trabajo isotérmico para la misma expansión.
4. **Interpretar físicamente**:
   - ¿Por qué el trabajo adiabático es **menor** que el trabajo isotérmico?
   - ¿Qué pasa con la temperatura del gas durante la expansión adiabática?

**Solución**

**1. Cálculo del Trabajo Adiabático \( W \)**
Primero, calculamos \( P_2 \) usando la relación adiabática:
\[
P_2 = P_1 \left( \frac{V_1}{V_2} \right)^\gamma \approx 1.892 \, \text{atm}
\]
Ahora, calculamos \( W \):
\[
W = \frac{P_1 V_1 - P_2 V_2}{\gamma - 1} \approx 616.2 \, \text{J}
\]

**Resultado:**
\[
\boxed{W \approx 616.2 \, \text{J}}
\]

`````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Parámetros
P1 = 10.0      # atm (presión inicial)
V1 = 1.0       # L (volumen inicial)
V2 = 4.0       # L (volumen final)
gamma = 7/5    # Coeficiente adiabático (gas diatómico)

# Constante adiabática
const = P1 * V1**gamma

# Volúmenes para la gráfica
V = np.linspace(V1, V2, 500)

# Presión P(V) = const / V^gamma
P = const / V**gamma

# Trabajo W (área bajo la curva)
W, _ = quad(lambda V: const / V**gamma, V1, V2)
W_J = W * 101.325  # Convertir L·atm a J
print(f"Trabajo adiabático W = {W:.2f} L·atm = {W_J:.2f} J")

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(V, P, label="\$P(V) = \\frac{P_1 V_1^\\gamma}{V^\\gamma}\$", color="blue")
plt.fill_between(V, P, alpha=0.3, color="blue", label="Área = Trabajo $W$")
plt.xlabel("Volumen $V$ (L)")
plt.ylabel("Presión $P$ (atm)")
plt.title("Trabajo en una Expansión Adiabática de un Gas Ideal")
plt.legend()
plt.grid(True)
plt.show()
`````
