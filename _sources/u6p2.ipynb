## Integral indefinida

Se define la función $G(x) = \int_a^x{f(t)\,dt}$. Notamos que $G$ es una función en donde al variar el valor del límite superior de la integral, cambia el valor de la integral. Calculemos dicha función para varios ejemplos:

## Función lineal

Considere $f(t) = t$

$G(x) = \int_0^x{t\,dt}$, con $x \in [0, 5]$

¿Qué función es?

```python
import numpy as np

def f(t):
    return t
x = np.linspace(0, 5, 50)
Gx = []
#integración de f(t)
for xi in x:
    #Extremos del intervalo
    a = 0
    b = xi
    #Número de subintervalos
    n = 100
    #Tamaño de la partición
    dt = (b-a)/n
    #Puntos de la partición
    t = np.linspace(a,b,n+1)
    t_med = (t[0:n] + t[1:n+1])/2 #t medios
    A_med = f(t_med) * dt #medias
    #Suma de Riemann
    Suma = sum(A_med)
    Gx.append(Suma)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,Gx)
plt.scatter(x,f(x))
```

G(x) = \int_{a}^{x}{f(t)\,dt}

tomamos valores distintos del límite inferior.
