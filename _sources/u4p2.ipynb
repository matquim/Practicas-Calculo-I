# Derivadas de funciones

**Objetivo**

Comparar derivadas analíticas con aproximaciones numéricas.

**Ejemplo**

```python
import math

def f(x):
    return math.sin(x)

def df(x):
    return math.cos(x)

x = math.pi / 4
h = 1e-5

aprox = (f(x + h) - f(x)) / h
exacta = df(x)

print("Aprox:", aprox)
print("Exacta:", exacta)
```
