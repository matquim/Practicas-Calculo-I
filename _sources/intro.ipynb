# Prácticas Cálculo I

El presente libro digital tiene como objetivo ofrecer actividades computacionales que puedan ser usadas en un primer curso de Cálculo. En particular, la organización se apega al temario de la asignatura de [`Cálculo I (Clave 2101)`](https://quimica.unam.mx/wp-content/uploads/2022/02/1111-Calculo-I-CCQ.pdf) de la Facultad de Química (UNAM), tal que las actividades aquí mostradas sirvan como referencia para ser usadas en las clases.



La estructura de este libro busca favorecer la interactividad y el aprendizaje de conceptos. La presentación de conceptos y de código de referencia se destaca en bloques como el siguiente:

## Ejecución

El código propuesto está desarrollado en `Python`. Con fines didácticos, se sugiere ejecutarlo en `Google Colab`. Como alternativa, también puede ejecutarse directamente desde la misma página web utilizando el botón `Live Code`, al cual se accede mediante el ícono con forma de `nave espacial` situado en la esquina superior derecha de las páginas.

Recuerde que puede ejecutar cada celda usando las teclas **Shift + Enter**, el botón *run* en Live Code o el triángulo de *Play* en `Google Colab`.
