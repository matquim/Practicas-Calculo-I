# Práctica 1: Aproximación numérica de límites

**Objetivo**

Aproximar límites evaluando funciones en valores cercanos a un punto.

````{admonition} Concepto
:class: tip

El límite de una función $f(x)$ cuando $x$ tiende a $a$ se escribe:

\begin{equation}
\lim_{x \to a} f(x)
\end{equation}

y describe el comportamiento de la función cerca de $a$.

````

**Ejemplo**


`````python
def f(x):
    return (x**2 - 1)/(x - 1)

valores = [0.9, 0.99, 0.999, 1.001, 1.01, 1.1]

for x in valores:
    print(x, f(x))
`````

**Ejercicios**

a) Aproximar el límite de $f(x) = \frac{x^2 - 4}{x - 2}$ cuando $x \to 2$.

b) Evaluar la función en valores cercanos a 2 por la izquierda y por la derecha.
