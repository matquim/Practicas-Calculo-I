En un gas ideal, las partículas tienen una **distribución de velocidades** dada por la **distribución de Maxwell-Boltzmann**:

$$
f(v) = 4 \pi \left( \frac{m}{2 \pi k T} \right)^{3/2} v^2 e^{-\frac{m v^2}{2 k T}}
$$

Donde:
- $ f(v) $: Función de distribución de velocidades.
- $ m $: Masa de una partícula.
- $ k $: Constante de Boltzmann ($ 1.38 \times 10^{-23} \, \text{J/K} $).
- $ T $: Temperatura absoluta.
- $ v $: Velocidad de una partícula.

La **energía cinética** de una partícula es:

$$
E = \frac{1}{2} m v^2
$$

La **energía promedio** $ \langle E \rangle $ se calcula como:

$$
\langle E \rangle = \frac{\int_{0}^{\infty} E \cdot f(v) \, dv}{\int_{0}^{\infty} f(v) \, dv}
$$

**Enunciado**

1. Calcular la **energía promedio $ \langle E \rangle $** de una partícula en un gas ideal a temperatura $ T $, usando la distribución de Maxwell-Boltzmann.
2. **Demostrar** que $ \langle E \rangle = \frac{3}{2} k T $ mediante integración numérica.
3. **Graficar** la distribución de Maxwell-Boltzmann $ f(v) $ y la distribución de energías $ f(E) $ para un gas de nitrógeno ($ m = 28 \, \text{u} = 4.65 \times 10^{-26} \, \text{kg} $) a $ T = 300 \, \text{K} $.
4. **Interpretar físicamente**:
   - ¿Por qué la energía promedio depende **solo de la temperatura** y no de la masa de las partículas?
   - ¿Cómo se relaciona este resultado con la **teoría cinética de los gases**?

**Solución**

**1. Cálculo de la Energía Promedio $ \langle E \rangle $**

La energía promedio teórica para un gas ideal es:
$$
\langle E \rangle = \frac{3}{2} k T
$$

**2. Demostración Numérica de $ \langle E \rangle = \frac{3}{2} k T $**

Usamos **SciPy** para calcular las integrales numéricamente y verificar el resultado teórico.

`````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy.constants import Boltzmann as k_B  # Constante de Boltzmann (1.38e-23 J/K)

# Parámetros para el nitrógeno (N2)
m = 4.65e-26  # Masa de una molécula de N2 (kg)
T = 300.0     # Temperatura (K)
k = k_B       # Constante de Boltzmann (J/K)

# Constante para la distribución de Maxwell-Boltzmann
A = (m / (2 * np.pi * k * T)) ** (3/2)

# Función de distribución de velocidades f(v)
def f_v(v):
    return 4 * np.pi * A * v**2 * np.exp(-m * v**2 / (2 * k * T))

# Función de energía E(v)
def E(v):
    return 0.5 * m * v**2

# Numerador: Integral de E * f(v)
numerator, _ = quad(lambda v: E(v) * f_v(v), 0, np.inf)

# Denominador: Integral de f(v)
denominator, _ = quad(f_v, 0, np.inf)

# Energía promedio numérica
E_avg_num = numerator / denominator

# Energía promedio teórica
E_avg_theory = (3/2) * k * T

print(f"Energía promedio numérica: {E_avg_num:.6e} J")
print(f"Energía promedio teórica: {E_avg_theory:.6e} J")
print(f"Diferencia relativa: {abs(E_avg_num - E_avg_theory) / E_avg_theory * 100:.6f}%")
`````
