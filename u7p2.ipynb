**Contexto Teórico**
Para un gas ideal, el **cambio de energía interna $ \Delta U $** a volumen constante está dado por:
$$
\Delta U = \int_{T_1}^{T_2} n C_v \, dT
$$
Donde:
- $ C_v $: Capacidad calorífica molar a volumen constante.
- $ n $: Número de moles.

Si $ C_v $ es constante:
$$
\Delta U = n C_v (T_2 - T_1)
$$

**Instrucciones**
1. Un gas ideal con $ n = 3 \, \text{moles} $ y $ C_v = 20.8 \, \text{J/mol·K} $ (constante) se calienta de $ T_1 = 298 \, \text{K} $ a $ T_2 = 350 \, \text{K} $ a volumen constante. Calcular el **cambio de energía interna $ \Delta U $**.
2. Si $ C_v $ varía con la temperatura como $ C_v(T) = a + bT $ (con $ a = 15.0 \, \text{J/mol·K} $ y $ b = 0.02 \, \text{J/mol·K}^2 $), recalcular $ \Delta U $.
3. **Graficar** $ C_v(T) $ y comparar el cambio de energía interna para ambos casos.
4. **Interpretar físicamente**:
   - ¿Por qué $ \Delta U $ es **positivo** en este proceso?
   - ¿Cómo afecta la dependencia de $ C_v $ con $ T $ al valor de $ \Delta U $?

**Solución**

**1. $ C_v $ Constante**
Usamos la integral para $ \Delta U $:
$$
\Delta U = n C_v (T_2 - T_1) = 3 \times 20.8 \times (350 - 298) = 3244.8 \, \text{J}
$$

**Resultado:**
$$
\boxed{\Delta U = 3244.8 \, \text{J}}
$$

**2. $ C_v(T) = a + bT $**
Usamos la integral para $ \Delta U $:
$$
\Delta U = n \int_{T_1}^{T_2} (a + bT) \, dT = n \left[ a T + \frac{b T^2}{2} \right]_{T_1}^{T_2}
$$
Sustituyendo los valores:
$$
\Delta U = 3 \left[ 15.0 \times (350 - 298) + \frac{0.02}{2} \times (350^2 - 298^2) \right] \approx 3350.88 \, \text{J}
$$

**Resultado:**
$$
\boxed{\Delta U = 3350.88 \, \text{J}}
$$

`````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Parámetros
n = 3.0          # Moles
T1 = 298.0       # K (temperatura inicial)
T2 = 350.0       # K (temperatura final)
Cv_const = 20.8  # J/mol·K (Cv constante)
a = 15.0         # J/mol·K (coeficiente de Cv(T))
b = 0.02         # J/mol·K^2 (coeficiente de Cv(T))

# Rango de temperaturas para la gráfica
T = np.linspace(T1, T2, 500)

# Cv constante
Cv_const_func = np.full_like(T, Cv_const)

# Cv(T) = a + bT
Cv_T = a + b * T

# Delta U para Cv constante
Delta_U_const = n * Cv_const * (T2 - T1)

# Delta U para Cv(T)
Delta_U_T, _ = quad(lambda T: n * (a + b * T), T1, T2)

# Gráfica
plt.figure(figsize=(12, 6))

# Gráfica de Cv(T)
plt.subplot(1, 2, 1)
plt.plot(T, Cv_const_func, label="\$C_v = 20.8 \, \text{J/mol·K}\$", color="blue")
plt.plot(T, Cv_T, label="\$C_v(T) = 15.0 + 0.02T \, \text{J/mol·K}\$", color="orange")
plt.xlabel("Temperatura $T$ (K)")
plt.ylabel("Capacidad calorífica $C_v$ (J/mol·K)")
plt.title("Capacidad Calorífica $C_v(T)$")
plt.legend()
plt.grid(True)

# Comparación de Delta U
plt.subplot(1, 2, 2)
plt.bar(["$C_v$ constante", "$C_v(T)$"], [Delta_U_const, Delta_U_T],
        color=["blue", "orange"], alpha=0.7)
plt.ylabel("Cambio de energía interna $\\Delta U$ (J)")
plt.title("Comparación de $\\Delta U$")
plt.grid(True)

plt.tight_layout()
plt.show()

print(f"ΔU (Cv constante) = {Delta_U_const:.2f} J")
print(f"ΔU (Cv(T)) = {Delta_U_T:.2f} J")
`````
