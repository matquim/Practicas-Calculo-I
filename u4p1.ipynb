# Aproximación numérica de la derivada

**Objetivo**

Aproximar la derivada de una función mediante el cociente incremental.

````{admonition} Concepto
:class: tip

La derivada de una función $f(x)$ en un punto $x$ se define como:

\begin{equation}
f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}
\end{equation}

````

El siguiente código aproxima numéricamente el valor de la derivada de $f(x)$ en $x=2$. Note que:

- Se define la función, $f(x)$
- Definimos el punto donde evaluar la derivada, $x$
- Definimos el tamaño de paso, $h$
- Evaluamos la fórmula

`````python
def f(x):
    return x**2

x = 2
h = 0.001

derivada = (f(x + h) - f(x)) / h
print(derivada)
`````

En este caso tan sencillo
\begin{equation}
f(x) = x^2
\end{equation}

sabemos que
\begin{equation}
\frac{df(x)}{dx} = 2x
\end{equation}

por lo que
\frac{df(2)}{dx} = 2(2) = 4.
