# Aproximación numérica de la derivada

**Objetivo**

Aproximar la derivada de una función mediante el cociente incremental.

````{admonition} Concepto
:class: tip

La derivada de una función $f(x)$ en un punto $x$ se define como:

\begin{equation}
f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}
\end{equation}

````

`````python
def f(x):
    return x**2

x = 2
h = 0.001

derivada = (f(x + h) - f(x)) / h
print(derivada)
`````
