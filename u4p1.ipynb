# Aproximación numérica de la derivada

## Definición de derivada

**Objetivo**

Aproximar la derivada de una función mediante el cociente incremental.

````{admonition} Concepto
:class: tip

La derivada de una función $f(x)$ en un punto $x$ se define como:

\begin{equation}
f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}
\end{equation}

````

El siguiente código aproxima numéricamente el valor de la derivada de
\begin{equation}
f(x) = x^2
\end{equation}

en $x=2$. Note que:

- Se define la función, $f(x)$
- Definimos el punto donde evaluar la derivada, $x$
- Definimos el tamaño de paso, $h$
- Evaluamos la fórmula

`````python
def f(x):
    return x**2

x = 2
h = 0.001

derivada = (f(x + h) - f(x)) / h
print(derivada)
`````

En este caso tan sencillo, sabemos que
\begin{equation}
\frac{df(x)}{dx} = 2x
\end{equation}

por lo que
\begin{equation}
\frac{df(2)}{dx} = 2(2) = 4.
\end{equation}

## Comparación entre derivada analítica y numérica para varias funciones

**Objetivo**

Comparar derivadas analíticas con aproximaciones numéricas.

Podemos generalizar la idea anterior a otros tipos de funciones. Por ejemplo, el siguiente código usa la librería `math` para definir

\begin{equation}
f(x) = \sin(x)
\end{equation}

así como su derivada analítica:
\begin{equation}
\frac{df(x)}{dx} = \cos(x)
\end{equation}

luego evalúa la derivada numérica:

~~~
import math

def f(x):
    return math.sin(x)

def df(x):
    return math.cos(x)

x = math.pi / 4
h = 1e-5

aprox = (f(x + h) - f(x)) / h
exacta = df(x)

print("Aprox:", aprox)
print("Exacta:", exacta)
~~~

Copie y pegue el código en la siguiente celda:

`````python
# Escriba código aquí
`````

`````python
import math

def f(x):
    return math.sin(x)

def df(x):
    return math.cos(x)

x = math.pi / 4
h = 1e-5

aprox = (f(x + h) - f(x)) / h
exacta = df(x)

print("Aprox:", aprox)
print("Exacta:", exacta)
`````
