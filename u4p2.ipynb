# Práctica 2: Límites laterales

````{admonition} Concepto
:class: tip

Los límites laterales se definen como:

\begin{equation}
\lim_{x \to a^-} f(x), \quad \lim_{x \to a^+} f(x)
\end{equation}
````

**Ejemplo**

`````python
def f(x):
    if x < 0:
        return -1
    else:
        return 1

valores = [-0.1, -0.01, 0.01, 0.1]

for x in valores:
    print(x, f(x))
`````
