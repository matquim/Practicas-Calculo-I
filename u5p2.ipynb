# Puntos Críticos

## Criterio de la Primera Derivada

````{admonition} Concepto
:class: tip

Los puntos críticos de una función son aquellos donde la primera derivada se vuelve cero. Estos pueden clasificarse según el cambio de signo de la primera derivada en la vecindad del punto crítico:
- Máximo: (+) -> (-)
- Mínimo: (-) -> (+)
````

Considere la función:
\begin{equation}
f(x) = x^3 - 3x^2
\end{equation}

la primera derivada de la función es:
\begin{equation}
f'(x) = 3x^2 - 6x = x(3x - 6)
\end{equation}

la que se hace cero cuando:
- $x = 0$
- $x = 2$

los cuales son los puntos críticos de la función.

Realice la gráfica de esta función en el intervalo $x \in [-1, 3]$, señalando los puntos críticos:

`````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 3, 100)
y = x**3 - 3*x**2

plt.plot(x,y)
plt.scatter([0, 2], [0, -4], color='red', label='Puntos críticos')
`````

En la siguiente celda:
- Defina un $h$ pequeño
- Para cada punto crítico: evalúe el valor de la derivada un poco antes y un poco después
- Use los cambios de signo para caracterizar los puntos críticos como máximos o mínimos.


`````python
def df(x):
    return 3*x**2 - 6*x

h = 0.01

print("==== Primer Punto Crítico ====")
print("x = 0")

print("Antes del punto crítico:", df(0 - h))
print("Punto Crítico: =", df(0))
print("Después del punto crítico:", df(0 + h))
print("Conclusión: Máximo")

print()

print("==== Segundo Punto Crítico ====")
print("x = 2")

print("Antes del punto crítico:", df(2 - h))
print("Punto Crítico: =", df(2))
print("Después del punto crítico:", df(2 + h))
print("Conclusión: Mínimo")
`````

## Criterio de la Segunda derivada

La segunda derivada proporciona un procedimiento más práctico para caracterizar los puntos críticos, ya que solo hay que evaluar directamente la segunda derivada en el punto crítico y evaliar el signo:
- Máximo: (-)
- Mínimo: (+)
- Punto de Inflexión: Cero

Regresando al ejemplo de $f(x) = x^3 - 3x^2$, 

la segunda derivada es:
\begin{equation}
f''(x) = 6x - 6
\end{equation}

`````python
def d2f(x):
    return 3*x**2 - 6*x

print("==== Primer Punto Crítico ====")
print("x = 0")

print("Punto Crítico: =", d2f(0))
print("Conclusión: Máximo")
print()

print("==== Segundo Punto Crítico ====")
print("x = 2")

print("Punto Crítico: =", d2f(2))
print("Conclusión: Mínimo")
`````

Además, $f''(x) = 6x-6$ se hace cero cuando $x=1$, este es un `punto de inflexión`, el cual indica un cambio en la concavidad de la curvatura de cóncavo a convexo.

Note que este punto de inflexión solo pudo ser detectado con la segunda derivada y no con la primera derivada.
