# Puntos Críticos

## Máximos y Mínimos

El criterio de la primera y segunda derivada permite identificar los máximos y mínimos de una función.

```python
import sympy

x = sympy.Symbol('x')

f_x = x**3 - 6*x**2 + 9*x + 1
print(f"Función: f(x) = {f_x}")

f_prime_x = sympy.diff(f_x, x)
print(f"Primera derivada: f'(x) = {f_prime_x}")

critical_points = sympy.solve(f_prime_x, x)
print(f"Puntos críticos: {critical_points}")

f_double_prime_x = sympy.diff(f_prime_x, x)
print(f"Segunda derivada: f''(x) = {f_double_prime_x}")

for point in critical_points:
    second_deriv_at_point = f_double_prime_x.subs(x, point)
    if second_deriv_at_point < 0:
        print(f"En x = {point}: f''(x) = {second_deriv_at_point} (Máximo local). Valor de la función: {f_x.subs(x, point)}")
    elif second_deriv_at_point > 0:
        print(f"En x = {point}: f''(x) = {second_deriv_at_point} (Mínimo local). Valor de la función: {f_x.subs(x, point)}")
    else:
        print(f"En x = {point}: f''(x) = {second_deriv_at_point} (Criterio no concluyente. Se necesita un análisis adicional).")
```

## Puntos de Inflexión

```python
import sympy

x = sympy.Symbol('x')
f_x = x**3 - 6*x**2 + 9*x + 1
f_double_prime_x = sympy.diff(f_x, x, 2)
print(f"Función: f(x) = {f_x}")
print(f"Segunda derivada: f''(x) = {f_double_prime_x}")

inflection_candidates = sympy.solve(f_double_prime_x, x)
print(f"Candidatos a puntos de inflexión (f''(x) = 0): {inflection_candidates}")
```
